Before we begin analyzing to see which region contains the most amount of playground equipment, we need to first import all our APIs and for formatting purposes, set our default resource configuration value.

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcdefaults()
```

Alright, now that that's taken care of, let us reference what the city of Pittsburgh looks like. This way we have a good general idea of the map we're working with.

```python
neighborhoods = gpd.read_file("Neighborhoods/Neighborhoods_.shp")
neighborhoods.plot();
```

Wow! Pretty neat, isn't it? Nonetheless, it is quite difficult to distinctly tell each neighborhood apart. Perhaps we should try to clearly identify each neighborhood section before continuing.

```python
neighborhoods.plot(edgecolor="black");
```

Much better! Okay, now that we can see each distinct neighborhood section, let's find how much playground equipment each individual neighborhood has.

```python
playground_equipment = pd.read_csv("Playground Equipment.csv")
playground_equipment_counts = playground_equipment["neighborhood"].value_counts()
playground_equipment_counts.index.name = "Neighborhood"
print(playground_equipment_counts.to_string(index=True))
```

Interesting! We can see that Squirrel Hill South appears to be the best neighborhood for my sub-metric. However, this neighborhood being the highest for my sub-metric does not guarantee it will be the highest for our combined metric. Anyway, for now, we need to standardize our data.

```python
playground_equipment_scale = playground_equipment_counts / 3.2
print(playground_equipment_scale.to_string(index=True))
```

Nice! Since we all agreed to do a 10-point scale, once all our data is gathered, we can combine them with some algebra to then see which neighborhood scores the highest. This'll then determine the answer to our metric; that is, what is the best neighborhood to raise a child in. Regardless, we need to interpret our data, so let's create a bar and a pie chart.

```python
playground_equipment_counts.plot.bar()

plt.figure()
playground_equipment_counts.plot.pie();
```

Yikes, just look at that mess! Because we have too many neighborhoods to analyze, instead, we should analyze how many neighborhoods have "x" amount of playground equipment.

```python
neighborhood_counts = playground_equipment_counts.value_counts().sort_index()

neighborhood_counts.plot.bar()

plt.figure()
neighborhood_counts.plot.pie();
```

Well, it sort of works. The only thing that is easy to perceive is that the most common amount of playground equipment in neighborhoods seems to be 3. Still, everything is hard to perceive as the axes should be reversed for the bar chart, and the pie chart doesn't even have any percentages. Let's fix that, shall we?

```python
plt.style.use("ggplot")
neighborhood_counts.plot.barh(figsize = (9.5, 9.5))
plt.title("x Neighborhoods With y Playground Equipment", y = 1.005)
plt.xlabel("Amount of Neighborhoods")
plt.ylabel("Number of Playground Equipment", labelpad = 5)
plt.tight_layout()

plt.figure()
plt.style.use("Solarize_Light2")
neighborhood_counts.plot.pie(figsize = (10, 10), autopct = "%.2f%%", fontsize = 6, wedgeprops={"edgecolor": "black"})
plt.title("Percentage of Neighborhoods With x Playground Equipment", y = 0.955)
plt.ylabel(None)
plt.tight_layout()

plt.rcdefaults()
```

Fascinating! Look how good they look. Now, for example, if we wanted to discover how many neighborhoods have 5 pieces of playground equipment, we can observe from the charts. From the bar chart, we can see that 9 neighborhoods have 5 pieces of playground equipment; this is 13.24% of every neighborhood in the dataset, which was figured out from the pie chart. In any case, before we finish our analysis, let's find the median and mean amounts of playground equipment that neighborhoods have.

```python
print("Mean:", playground_equipment_counts.median())
print("Median:", playground_equipment_counts.mean())
```

So, now we know the mean is 5, and the median is 6.544117647058823 for the amount of playground equipment that neighborhoods have. Well, that's enough examination. Before we wrap up, let's visualize what each neighborhood area on the map received for a rating score in the city of Pittsburgh map.

```python
playground_equipment_map = neighborhoods.merge(playground_equipment_scale, how = "left", left_on = "hood", right_on = "Neighborhood")
playground_equipment_map.plot(column = "neighborhood", legend = True);
```

Hmmm... that's not looking too great. For me, at least, it's quite demanding to even keep looking at this map. That only means we have to give the map a good makeover, right?

```python
playground_equipment_map.plot(column = "neighborhood",
                              cmap = "Greens",
                              edgecolor = "white",
                              legend = True,
                              legend_kwds = {"label": "Amount of Playground Equipment"},
                              figsize = (10, 6),
                              missing_kwds = {"color": "#FFD580"}
                             )
plt.title("Amount of Playground Equipment in Each Pittsburgh Neighborhood");
```

Just look at that! Now, we can properly visualize the rating that each neighborhood received geographically. The darker the shade of green, the more highly it scored for my sub-metric, and light orange areas did not have playground equipment. Well, at any rate, that's all that I got, bye now!
